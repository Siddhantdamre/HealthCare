```python
import re
import pandas as pd
import spacy
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import folium
from geopy.geocoders import Nominatim
import torch
from transformers import pipeline  # For NER

nltk.download('stopwords')
nltk.download('wordnet')
nlp = spacy.load("en_core_web_sm")  # Or a more specific medical model

# --- Load your data ---
# This will depend on how your medical reports are stored (e.g., CSV, text files, database)
# Example (for CSV):
# reports_df = pd.read_csv("your_reports.csv")
# Ensure your CSV has a 'text' column containing the report text
# reports_df = pd.DataFrame({'text': ["Patient has symptoms of fever and cough. Pincode: 123456",
#                                     "Confirmed case of influenza. Pincode: 654321",
#                                     "Report of myocardial infarction. Pincode: 987654"]})  # Demo data

# --- Use the sample dataset ---
sample_data = {
    'report_id': [1, 2, 3, 4, 5, 6, 7],
    'text': [
        "Patient John Doe, pincode 600001, presents with fever, cough, and sore throat. Possible influenza.",
        "Report from clinic in pincode 600002: Two cases of gastroenteritis reported. Patient pincode 600002, Patient pincode 600005.",
        "Confirmed case of pneumonia in pincode 600003. Patient had a history of asthma.",
        "Outbreak of conjunctivitis in pincode 600004. Please take necessary precautions.",
        "Pincode 600001: Suspected case of chickenpox in a child. Pincode: 600007",
        "Report from pincode 600006 indicates a rise in dengue fever cases. Pincode: 600006",
        "Patient in pincode 600007 diagnosed with hypertension and diabetes. Pincode: 600007"
    ],
    'pincodes': [
        "600001",
        "600002 600005",
        "600003",
        "600004",
        "600001 600007",
        "600006",
        "600007"
    ]
}

reports_df = pd.DataFrame(sample_data)

# Phase 2: Text Preprocessing and Entity Extraction

def preprocess_text(text):
    text = re.sub(r'[^\w\s]', '', text)
    text = text.lower()
    tokens = text.split()
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return ' '.join(tokens)

reports_df['processed_text'] = reports_df['text'].apply(preprocess_text)

# --- Disease Entity Extraction ---
# Using spaCy for NER (can be replaced with a medical NER model)
def extract_diseases_spacy(text):
    doc = nlp(text)
    diseases = [ent.text for ent in doc.ents if ent.label_ in ["DISEASE", "SYMPTOM"]]  # Adjust labels as needed
    return diseases

reports_df['diseases'] = reports_df['processed_text'].apply(extract_diseases_spacy)

# --- Pincode Extraction ---
def extract_pincodes(text):
    # This is a basic pincode regex (Indian format)
    pincode_pattern = r"\b\d{6}\b"
    pincodes = re.findall(pincode_pattern, text)
    return pincodes

# reports_df['pincodes'] = reports_df['text'].apply(extract_pincodes) #Original

# --- Adjusted pincode handling ---
def get_pincode_list(pincode_string):
    return pincode_string.split()

# reports_df['pincodes'] = reports_df['text'].apply(extract_pincodes)
# reports_df['pincodes'] = reports_df['pincodes'].apply(lambda x: [str(item) for item in x])
reports_df['pincodes'] = reports_df['pincodes'].apply(get_pincode_list)
# reports_df['pincodes_str'] = reports_df['pincodes'].apply(lambda x: ' '.join(x))

# Phase 3: Terminology Mapping (Illustrative)

# --- Simple Disease Mapping Dictionary ---
# In a real application, this would be replaced with a medical ontology or more sophisticated mapping
disease_mapping = {
    "myocardial infarction": "Heart Attack",
    "heart attack": "Heart Attack",
    "influenza": "Flu",
    "flu": "Flu",
    "gastroenteritis": "Gastroenteritis",
    "pneumonia": "Pneumonia",
    "asthma": "Asthma",
    "conjunctivitis": "Conjunctivitis",
    "chickenpox": "Chickenpox",
    "dengue fever": "Dengue Fever",
    "hypertension": "Hypertension",
    "diabetes": "Diabetes",
    "fever": "Fever",
    "cough": "Cough",
    "sore throat": "Sore Throat",
    "symptoms": "Symptoms",
    "possible": "Possible"
}

def map_diseases(diseases, mapping):
    mapped_diseases =
    for disease in diseases:
        mapped_diseases.append(mapping.get(disease.lower(), disease)) # Use .lower() for case-insensitive matching
    return mapped_diseases

reports_df['mapped_diseases'] = reports_df['diseases'].apply(map_diseases, args=(disease_mapping,))

# Phase 4: Data Aggregation and Geocoding

# --- Data Aggregation ---
disease_pincode_counts = {}
for index, row in reports_df.iterrows():
    for disease in row['mapped_diseases']:
        for pincode in row['pincodes']: #Access pincode from list
            if pincode not in disease_pincode_counts:
                disease_pincode_counts[pincode] = {}
            disease_pincode_counts[pincode][disease] = disease_pincode_counts[pincode].get(disease, 0) + 1

# --- Geocoding ---
# This part requires an internet connection and may have usage limits.
geolocator = Nominatim(user_agent="disease_mapping")  # Replace "disease_mapping" with your app name

def get_coordinates(pincode):
    try:
        location = geolocator.geocode(f"pincode {pincode}, India", exactly_one=True)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except Exception as e:
        print(f"Error geocoding pincode {pincode}: {e}")
        return None, None

pincode_coordinates = {}
for pincode in disease_pincode_counts.keys():
    lat, lon = get_coordinates(pincode)
    pincode_coordinates[pincode] = {'lat': lat, 'lon': lon}

# Phase 5: Heatmap Generation

# --- Heatmap Generation ---
m = folium.Map(location=, zoom_start=6)  # Center of India

for pincode, counts in disease_pincode_counts.items():
    coords = pincode_coordinates[pincode]
    if coords['lat'] is not None and coords['lon'] is not None:
        popup_text = f"Pincode: {pincode}<br>"
        for disease, count in counts.items():
            popup_text += f"{disease}: {count}<br>"
        folium.CircleMarker(
            location=[coords['lat'], coords['lon']],
            radius=10,  # Adjust radius as needed
            popup=popup_text,
            color='red',
            fill=True,
            fill_color='red',
            fill_opacity=0.6
        ).add_to(m)

m.save("disease_heatmap.html")  # Save the heatmap to an HTML file

# Phase 6: Report Generation (Illustrative)

# --- Report Generation (Basic) ---
def generate_report(disease_pincode_counts):
    report = "Disease Outbreak Report\n\n"
    for pincode, counts in disease_pincode_counts.items():
        report += f"Pincode: {pincode}\n"
        if not counts:
            report += "  No diseases reported.\n"
        else:
            for disease, count in counts.items():
                report += f"  - {disease}: {count} cases\n"
        report += "\n"
    return report

report = generate_report(disease_pincode_counts)
print(report)

# --- Planned Course of Action (This is a complex part and requires external knowledge) ---
# For a basic example, you could have a dictionary mapping diseases to general recommendations
disease_recommendations = {
    "Heart Attack": "Increase public awareness, promote healthy lifestyle.",
    "Flu": "Encourage vaccination, practice good hygiene.",
    "Gastroenteritis": "Promote hygiene, ensure clean water supply",
    "Pneumonia": "Administer antibiotics, provide respiratory support",
    "Asthma": "Avoid triggers, use inhalers",
    "Conjunctivitis": "Practice hygiene, use eye drops",
    "Chickenpox": "Isolate patients, administer antiviral medication",
    "Dengue Fever": "Control mosquito breeding, provide supportive care",
    "Hypertension": "Recommend lifestyle changes, prescribe medication",
    "Diabetes": "Manage diet, exercise, and medication",
    # ... add more diseases and recommendations
}

def suggest_actions(disease_pincode_counts, disease_recommendations):
    actions = {}
    for pincode, counts in disease_pincode_counts.items():
        actions[pincode] =
        for disease, count in counts.items():
            if disease in disease_recommendations:
                actions[pincode].append(disease_recommendations[disease])
    return actions

actions = suggest_actions(disease_pincode_counts, disease_recommendations)

print("\nPlanned Actions:")
for pincode, action_list in actions.items():
    print(f"Pincode {pincode}:")
    if action_list:
        for action in action_list:
            print(f"  - {action}")
    else:
        print("  No specific actions suggested.")
```
